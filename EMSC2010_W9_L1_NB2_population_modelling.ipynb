#### The logistic population model:

$x_{\mathrm{next}} = a x_{\mathrm{initial}}(1-x_{\mathrm{initial}})$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size
* $x_{\mathrm{next}}$ is next year's population size
* $a$ is the ability to reproduce.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Code the logistic population model

#values
a = 2.4
x_initial = 0.7

#equation
x_next = a*x_initial*(1-x_initial)
print(x_next)
```

- **for-loop**
- python will repeat (loop) an equation over and over again for a specific amount of time (e.g. 10, until it reaches 10)
- good for population modelling - e.g. want to know the population after 50 generations

```python
#first for-loop example

a = 2.4
for i in range(10):
  print(i)
```

```python
#let's try it out

a = 1.11
x = np.zeros(50) #50 generations
x[0] = 0.5 #first population entry, address 0  #this is our starting population
for i in range (1, 50):   #all population entries from addresses 1 to 50
  x[i] = a*x[i-1]*(1-x[i-1])  #logistic model equation
  #indent indicates it is part of the loop

plt.plot(x)
plt.title('Logistic population model')
plt.xlabel('Generations') #x axis = generations (repeats)
plt.ylabel('Population size') #y axis = population size between 0-1
plt.ylim(0, 1) #y axis limits
plt.xlim(0,50) #x axis limits
plt.show()
```

#### The logistic population model with harvesting:

$x_{\mathrm{next}} = a x_{\mathrm{initial}}(1-x_{\mathrm{initial}}) - h$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size.
* $x_{\mathrm{next}}$ is next year's population size.
* $a$ is the ability to reproduce.
* $h$ is the size of the harvest between 0 and 1.


**harvesting = population control? Consistent removal of an amount of the population**


e.g. subtracting 0.1 from the population each year
- understanding population dynamics - demonstrates critical thresholds where small/large amounts of harvest (population removal) has *drastic effects on the population*
- Small changes, big outcomes in a simple model
- this is an **abstraction** - simple model guess

```python
### Code the logistic population model with harvesting

a = 1.11
h = 0.025
x = np.zeros(50) #50 generations
x[0] = 0.5 #first population entry, address 0  #this is our starting population

for i in range (1, 50):   #all population entries from addresses 1 to 50
  x[i] = a*x[i-1]*(1-x[i-1]) - h  #logistic model equation with harvesting
  x[i] = np.maximum(x[i],0) #compares x[i] with 0, limits -  makes sure we don't have negative population values
  #indent indicates it is part of the loop


plt.plot(x)
plt.title('Logistic population model')
plt.xlabel('Generations') #x axis = generations (repeats)
plt.ylabel('Population size') #y axis = population size between 0-1
plt.ylim(0, 1) #y axis limits
plt.xlim(0,50) #x axis limits
plt.show()
```

#### The logistic population model with randomness:

Random environmental cariations will modify the ability to reproduce - influence the population size. *using randomness*

$x_{\mathrm{next}} = (a+r) x_{\mathrm{initial}}(1-x_{\mathrm{initial}})$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size.
* $x_{\mathrm{next}}$ is next year's population size.
* $a$ is the ability to reproduce.
* **$r$ is a normally distributed random number that changes each year.**


we control r to define how much influence the random number has on our equation/population model. Adjusting reproducability (a) each year with a bit of randomness.
- adding (a+r) to the equation

The bigger the standard deviation e.g.0.9 the more variability/randomness will be in the model (large variation of a/reproducability)

```python
### Code the logistic population model with randomness

a = 2.4
h = 0.0
x = np.zeros(50) #50 generations
x[0] = 0.5 #first population entry, address 0  #this is our starting population

for i in range (1, 50):   #all population entries from addresses 1 to 50
  r = np.random.randn()*0.4 #0.4 is the standard deviation (we control this to limit r)
  x[i] = (a+r)*x[i-1]*(1-x[i-1]) - h  #logistic model equation with harvesting
  x[i] = np.maximum(x[i],0) #compares x[i] with 0, limits -  makes sure we don't have negative population values
  #indent indicates it is part of the loop


plt.plot(x)
plt.title('Logistic population model')
plt.xlabel('Generations') #x axis = generations (repeats)
plt.ylabel('Population size') #y axis = population size between 0-1
plt.ylim(0, 1) #y axis limits
plt.xlim(0,50) #x axis limits
plt.show()
```
